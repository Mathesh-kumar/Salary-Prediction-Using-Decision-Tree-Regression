## **Machine Learning model to predict salary for new employee based on position of work using decision tree regression.**
  1. Importing needed libraries
  2. Data preprocessing
  3. Training DTR model on dataset
  5. Predicting new results
  6. Visualizing DTR results

```
# 1. importing needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# 2. data preprocessing
# 2.1. importing needed dataset
df = pd.read_csv('Position_Salaries.csv')
df.head()
```

```
# 2.2. splitting dependent & independent variable
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```
# since our dataset is very small, we are not using train_test_split.
```

```
# 3. training dtr on the dataset.

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X, y)
```

```
# 4. predicting new result
# decision is useful for multi dimentional dataset. since our dataset is very small the prediction result in not effective.

dtr.predict([[6.5]])
```

```
# 6. visualizing decision tree regression result

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, dtr.predict(X_grid), color = 'blue')
plt.title('Decision Tree Regression results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
